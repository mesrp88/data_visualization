```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
url='/content/unrate.csv'
df= pd.read_csv(url)
df.head()
```

```
df.shape
```

```
df['DATE']= pd.to_datetime(df['DATE'])
```

```
df.info()
```

```
plt.plot()
plt.show()
```

```
plt.plot(df['DATE'][:12], df['VALUE'][:12])
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment rate')
plt.title('Unemployment in different month, 1948')
plt.show()
```

here I am creating multiplot

```
fig= plt.figure()
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
plt.show()
```

```
fig= plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
plt.show()
```

```
fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.plot(df['DATE'][:12], df['VALUE'][:12])
ax2.plot(df['DATE'][12:24], df['VALUE'][12:24])
plt.show()
```

```
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(df[0:12]['DATE'], df[0:12]['VALUE'])
ax1.set_title('Monthly Unemployment Rate, 1948')
ax2.plot(df[12:24]['DATE'], df[12:24]['VALUE'])
ax2.set_title('Monthly Unemployment Rate, 1949')
plt.show()
```

```
fig =plt.figure(figsize=(12,5))
for i in range(2):
    ax= fig.add_subplot(2,1,i+1)
    start_index=i*12
    end_index= (i+1)*12
    ax.plot(df['DATE'][start_index:end_index], df['VALUE'][start_index:end_index])
plt.show()
```

```
fig =plt.figure(figsize=(12,12))
for i in range(5):
    ax= fig.add_subplot(5,1,i+1)
    start_index=i*12
    end_index= (i+1)*12
    ax.plot(df['DATE'][start_index:end_index], df['VALUE'][start_index:end_index])
plt.show()
   
```

```
df['MONTH'] =df['DATE'].dt.month
color= ['red', 'blue']
fig=plt.figure(figsize=(6,3))
for i in range(2):
    start_index=i*12
    end_index=(i+1)*12
    plt.plot(df['MONTH'][start_index:end_index], df['VALUE'][start_index:end_index], c=color[i])
plt.show()
```

Let's visualize 5 years worth of unemployment rates on the same subplot.

```
fig=plt.figure(figsize=(10,6))
color=['red','blue','green', 'orange', 'black']
for i in range(5):
  start_index=i*12
  end_index=(i+1)*12
  plt.plot(df['MONTH'][start_index:end_index],df["VALUE"][start_index:end_index],
           c=color[i], label=str(1948+i))
plt.legend(loc='upper left')
plt.title('Monthly Unemployment Trends, 1948-1952')
plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')

plt.show()

```
